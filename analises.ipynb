```python
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity
```

```python
colunas = ['o3.ppb', 'co.ppb', 'so2.ugm3', 'pm25.ugm3', 'temperature', 'humidity', 'wind.speed', 'wind.direction', 'rainfall']
df_pca = df[colunas].dropna()

# Padronizar os dados (Z-score)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_pca)

# Teste de AdequaÃ§Ã£o para PCA
kmo_all, kmo_model = calculate_kmo(df_scaled)
bartlett_chi2, bartlett_p_value = calculate_bartlett_sphericity(df_scaled)

print(f"KMO: {kmo_model:.2f}")
print(f"Bartlett's Test: Chi2 = {bartlett_chi2:.2f}, p-value = {bartlett_p_value:.5f}")

# Verificar adequaÃ§Ã£o dos dados
if kmo_model < 0.50 or bartlett_p_value > 0.05:
    print("Os dados nÃ£o sÃ£o adequados para a PCA. Ajuste necessÃ¡rio.")
else:
    print("Os dados sÃ£o adequados para a PCA. Continuando...")

# Aplicar PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
explained_variance = pca.explained_variance_ratio_

# Plotar a biplot
fig, ax = plt.subplots(figsize=(5, 4))

# Adicionar o cÃ­rculo guia
circle = plt.Circle((0, 0), 1, color='grey', fill=False, linestyle='--')
ax.add_artist(circle)

# Adicionar vetores das variÃ¡veis
scaling_factor = 1
for i, col in enumerate(colunas):
    ax.arrow(0, 0, pca.components_[0, i] * scaling_factor, pca.components_[1, i] * scaling_factor,
             color='red', alpha=0.8, head_width=0.05)
    plt.text(pca.components_[0, i] * scaling_factor * 1.1, pca.components_[1, i] * scaling_factor * 1.1,
             col, color='darkred', fontsize=12)

# ConfiguraÃ§Ãµes do grÃ¡fico
plt.xlabel(f'PC1 ({explained_variance[0]*100:.2f}%)')
plt.ylabel(f'PC2 ({explained_variance[1]*100:.2f}%)')
plt.axhline(0, color='grey', lw=0.8)
plt.axvline(0, color='grey', lw=0.8)
plt.title('Biplot - AnÃ¡lise de Componentes Principais (PCA) com PrÃ©-processamento Ajustado')
plt.grid()
plt.show()
```
